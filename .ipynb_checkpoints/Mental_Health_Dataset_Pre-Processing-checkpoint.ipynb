```python
#Loading and displaying the dataset
Mental_Health_Dataset = pd.read_csv('Mental Health Dataset.csv')
Mental_Health_Dataset.head()
```

```python
#Check the size of the dataset
#timing
start_time = time.time()

print(f'The dataset has {Mental_Health_Dataset.shape[0]} data objects')

#Checking the time taken
end_time = time.time()

# Calculate elapsed time
elapsed_time = end_time - start_time
print(f"\nTime taken: {elapsed_time:.4f} seconds")
```

```python
# Since the dataset is large, we want to focus on UK Health dataset Only
#timing
start_time = time.time()
#Get UK data only
UK_Mental_Health = Mental_Health_Dataset[Mental_Health_Dataset['Country']=='United Kingdom']

#Checking the time taken
end_time = time.time()

# Calculate elapsed time
elapsed_time = end_time - start_time
print(f"\nTime taken: {elapsed_time:.4f} seconds")
```

```python
UK_Mental_Health.head()
```

```python
print(f'The new dataset has {UK_Mental_Health.shape[0]} rows')
```

```python
# Starting with data cleaning; Checking for Null values, duplicates and Outliers

#Null check
Null_check = UK_Mental_Health.isnull().values.any()

if Null_check:
    print("There are Null values in the dataset.")
else:
    print("There are no Null values in the dataset.")

UK_Mental_Health.isnull().sum()
```

```python
#Percentage Check of null
Null_Percent = (UK_Mental_Health['self_employed'].isnull().sum()/UK_Mental_Health.shape[0])*100
print(f'The null percentage is {Null_Percent:.3f} %')
```

```python
#Since the null percentage is 1.54%, it is small meaning, removing them will not have a big impact on the data
UK_Mental_Health_NoNull = UK_Mental_Health.dropna()
null_check = UK_Mental_Health_NoNull.isnull().values.any()

if null_check:
    print("There are Null values in the dataset.")
else:
    print("There are no Null values in the dataset.")
```

```python
# Check for duplicates in rows
Duplicates = UK_Mental_Health_NoNull.duplicated().sum()

if Duplicates:
    print("There are duplicate rows in the dataset.")
else:
    print("There are no duplicate rows in the dataset.")

print(f'\nThe Total Number of duplicates are {Duplicates}')
```

```python
#Specific rows with duplicates
UK_Mental_Health_NoNull.loc[UK_Mental_Health_NoNull.duplicated(),:]
```

```python
# Percentage of Duplicated rows
total_rows = UK_Mental_Health_NoNull.shape[0]
Percent_duplicated = (UK_Mental_Health_NoNull.duplicated().sum()/total_rows)*100
print(f'The percentage of duplicated rows is: {Percent_duplicated:.2f}%')
```

```python
# We drop the duplicates
Non_duplicated_data = UK_Mental_Health_NoNull.drop_duplicates()
Duplicates = Non_duplicated_data.duplicated().sum()
if Duplicates:
    print("There are duplicate rows in the dataset.")
else:
    print("There are no duplicate rows in the dataset.")

print(f'\nThe Total Number of duplicates are {Duplicates}')

print(f'\nTotal number of rows in our dataset is: {Non_duplicated_data.shape[0]}')
```

```python
# we remove the Timestamp column given we will not use it for the analysis
Non_duplicated_data.drop('Timestamp', axis =1, inplace=True)
Non_duplicated_data.head()
```

```python
Cleaned_UK_Mental_Health = Non_duplicated_data
Cleaned_UK_Mental_Health.head()
```

```python
Total_rows = Cleaned_UK_Mental_Health.shape[0]
Total_columns = Cleaned_UK_Mental_Health.shape[1]
print(f'The dataset has {Total_rows} rows and {Total_columns} columns')
```

```python
# Our new Dataset is; we write that back to CSV for further Analysis

Cleaned_UK_Mental_Health.to_csv('Cleaned_UK_Mental_Health.csv', index=False)
```
